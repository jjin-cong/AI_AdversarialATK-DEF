```python
datasetnames = ["Meta", "Test", "Train"]
x_Meta = []
x_Test = []
x_Train = []
for datasetname in datasetnames:
  if datasetname == "Train": # Train 데이터는 43개 class 별로 디렉토리 한 번 더 있음
    dataset = []
    j = 0
    dirnums_unorder = os.listdir(datasetname)
    dirnums = natsort.natsorted(dirnums_unorder)
    # print(dirnums)
    for dirnum in dirnums:
      datasetsub = []
      k = 0
      path = os.getcwd() + "/Train/" + dirnum + "/"
      filenames_unorder = os.listdir(path)
      filenames = natsort.natsorted(filenames_unorder)
      # print(filenames)
      for filename in filenames:
        with open(os.path.join(path, filename), 'rb') as f: ## csv로 열 수 있나?
          file = f.read()
          datasetsub.insert(k,file)
          k += 1
      dataset.insert(j,datasetsub)
      j += 1
    x_Train = dataset
    print("x_Train complete")

  else:
    dataset = []
    i = 0
    filenames_unorder = os.listdir(datasetname)
    filenames = natsort.natsorted(filenames_unorder)
    for filename in filenames:
      with open(os.path.join(datasetname, filename), 'rb') as f: ## csv로 열 수 있나?
        file = f.read()
        dataset.insert(i,file)
        i += 1
    if datasetname == "Meta":
      x_Meta = dataset
      print("x_Meta complete")
    if datasetname == "Test":
      x_Test = dataset
      print("x_Test complete")
```
